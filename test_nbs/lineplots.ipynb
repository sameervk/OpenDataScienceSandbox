```python
import seaborn as sns
```

```python
df1 = sns.load_dataset("fmri")
df1
```

```python
sns.lineplot(data=df1, x="timepoint", y="signal", hue="region", size="event")
```

```python
sns.lineplot(data=df1, x="timepoint", y="signal", hue="region", style="event")
```

```python
sns.scatterplot(data=df1, x="timepoint", y="signal", hue="region", style="event")
```
